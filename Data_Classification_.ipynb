```python
from numpy import where #where is used to find indices of elements
from collections import Counter #used to count the number of times an element occus
from sklearn.datasets import make_blobs #create a dummy dataset
from matplotlib import pyplot
```

```python
#create the dataset
X,y=make_blobs(n_samples=1000,centers=2,random_state=1)
```

```python
#summarize dataset
print(X.shape,y.shape)
```

```python
#balance the data
counter=Counter(y)
print(counter)
```

```python
#plot the dataset
for label, _ in counter.items():
  #get the indices
  row_indices=where(y==label)[0]
  #plot scatterplot
  pyplot.scatter(X[row_indices,0],X[row_indices,1],label=str(label))
```
